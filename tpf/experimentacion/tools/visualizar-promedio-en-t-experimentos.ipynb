# Visualizar el promedio en t de los valores de cada grupo

```python
import matplotlib.pyplot as plt
import pandas as pd

# FROM https://stackoverflow.com/a/36416258/5948384
# path = r'C:\DRO\DCL_rawdata_files'                     # use your path
# all_files = glob.glob(os.path.join(path, "*.csv"))     # advisable to use os.path.join as this makes concatenation OS independent

all_files = ['groups_1.csv', 'groups_2.csv']  
df_from_each_file = (pd.read_csv(f) for f in all_files)
concatenated_df   = pd.concat(df_from_each_file, ignore_index=True)

# print(concatenated_df)

grouped_df = concatenated_df.groupby(['time'])


# print(grouped_df.mean())

# print(grouped_df.std())
```

```python
#print(concatenated_df)

a_vals = grouped_df.agg(["mean", "std"])['A'].reset_index()

#print(a_vals['time'])
# print(grouped_df.groups)
plt.rcParams["figure.figsize"] = (20,10)
plt.errorbar(a_vals['time'][:45], a_vals['mean'][:45], a_vals['std'][:45], linestyle='None', marker='^')

plt.show()
```
